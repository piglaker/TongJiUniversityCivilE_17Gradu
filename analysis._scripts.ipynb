```python
import pandas as pd
import re
```

```python
file = pd.read_excel("./档案去向核对(1).xlsx")
```

```python
data = file['收件单位名称'].tolist()
```

```python
#tju = ["地下系", "桥梁系", "建工系", "防灾系", "水利系"]
tju = []

#otherUniversity = ["大学"]
otherUniversity = []

#job = ["公司"]
job = []

#Unemployment = ["人才"]
Unemployment = []

home = {"TongJi":0, "OtherUniversity":0, "Job":0, "Unemployment":0, "Others":0, "Total":len(data)}

status = []

def go_otheruniversity(element):
    if re.search("大学|研究所"), element) and not re.search("同济", element):
        return True
    else:
        return False

for element in data:
    if not type(element) == type("string"):
        continue 

    if go_otheruniversity(element):
        home["OtherUniversity"] += 1
        otherUniversity.append(element)
    else:
        status.append(element)    

otherUniversity

```

```python

status_expand = []

def stay_CivilEnginnering(element):
    CEDict = ["地下系", "桥梁系", "建工系", "防灾系", "水利系", "学院"]
    for d in CEDict:
        if re.search(d, element):
            return True
    else:
        return False

def stay_inTJU(element):
    if stay_CivilEnginnering(element) or re.search("同济", element):
        return True
    else:
        return False

for element in status:
    if not type(element) == type("string"):
        continue 

    if stay_inTJU(element):
        home["TongJi"] += 1
        tju.append(element)
    else:
        status_expand.append(element)    

print(len(status_expand))
```

```python
status_expand_expand = []

def go_forjob(element):
    CEDict = ["公司", "集团"]
    for d in CEDict:
        if re.search(d, element):
            return True
    else:
        return False

def go_forJob(element):
    if go_forjob(element):
        return True
    else:
        return False

for element in status_expand:
    if not type(element) == type("string"):
        continue 

    if go_forJob(element):
        home["Job"] += 1
        job.append(element)
    else:
        status_expand_expand.append(element)

print(job)

home["Unemployment"] += len(status_expand_expand)
```
